Параметр linestyle:

```python
styles = ['-', '--', '-.', ':', 'steps']

for ls in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', ls=ls)
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Параметр marker:

```python
styles = ['+', ',', '.', '1', '2', '3', '4' ]

for marker in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', marker=marker, ls=':', mec='b')
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Контроль толщины линии графика осуществляется за счет параметра linewidth:

```python
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, c='cornflowerblue', linewidth=2)
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

## Диаграмма

Подробнее с примерами кода:
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar

```python
x = np.arange(5)
height = [20, 23, 25, 24, 26]

plt.bar(x, height, align='center', color='c', linewidth = 100)

plt.title('Продажи товара Х по месяцам')

plt.ylabel('млн руб.')
plt.xticks(x, ('Январь', 'Февраль', 'Март', 'Апрель', 'Май'))

plt.show()
```

## Heatmap

Более подробно с примерами кода:

https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html

```python
np.random.seed(42)
corr = np.random.rand(10,10)
corr = (corr.dot(corr.T) - corr.dot(corr.T).min()) / (corr.dot(corr.T).max() - corr.dot(corr.T).min())


plt.imshow(corr)
plt.colorbar()
plt.show()
```

```python
plt.imshow(corr, cmap='hot')
plt.colorbar()
plt.show()
```
